# Data Analysis and Visualizations

```python
placement_percent = df['Placement'].value_counts(normalize = True)
placement_percent
```

```python
print(placement_percent)
plt.figure(figsize = (5,4))
plt.title('Placement status')
plt.bar(['Not placement', 'Placement'], df['Placement'].value_counts(), edgecolor = 'k', width = 0.6)
plt.savefig("placement_status.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Convert Placement column to binary for correlation
df_corr = df.copy()
df_corr['Placement'] = df_corr['Placement'].map({'Yes': 1, 'No': 0})

# Compute correlation matrix with target
correlation = df_corr.corr(numeric_only=True)['Placement'].drop('Placement').sort_values(ascending=False)

# Plot correlation
plt.figure(figsize=(8, 5))
sns.barplot(x=correlation.values, y=correlation.index, palette='coolwarm')
# plt.title("Correlation of Features with Placement Outcome")
plt.title("Fectors are responsible for Placement Outcome")
plt.xlabel("Correlation with Placement")
plt.savefig("Fectors_placement.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
df_corr = df.copy()
df_corr['Placement'] = df_corr['Placement'].map({'Yes': 1, 'No': 0})
df_corr['Internship_Experience'] = df_corr['Internship_Experience'].map({'Yes': 1, 'No': 0})

# Calculate correlation
correlation_value = df_corr.corr(numeric_only=True)['Placement']['Internship_Experience']
print(f"Correlation between Internship Experience and Placement: {correlation_value:.4f}")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Internship_Experience', hue='Placement', palette='Set1')
plt.title("Placement vs Internship Experience")
plt.xlabel("Internship Experience")
plt.ylabel("Number of Students")
plt.savefig("Internship_Experience_placement.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
# Create a copy and convert 'Placement' to numeric
df_corr = df.copy()
df_corr['Placement'] = df_corr['Placement'].map({'Yes': 1, 'No': 0})

# Calculate correlation
correlation_value = df_corr.corr(numeric_only=True)['Placement']['Academic_Performance']
print(f"Correlation between Academic Performance and Placement: {correlation_value:.4f}")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Academic_Performance', hue='Placement', palette='Set1')
plt.title("Placement vs Academic Performance")
plt.xlabel("Academic Performance")
plt.ylabel("Number of Students")
plt.savefig("Academic_Performance.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
Not_Placement = df[df['Placement'] == 'No']

plt.figure(figsize = (6, 4))
sns.countplot(data = Not_Placement, x = 'Academic_Performance', order = sorted(Not_Placement['Academic_Performance'].unique()), palette = 'magma')
plt.title("Academic Performance of Not placed Students")
plt.xlabel("Academic Performance Score")
plt.ylabel("Number of Students")
plt.savefig("Academic_Performance1.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
# Filter students who were not placed
Placement = df[df['Placement'] == 'Yes']

# Plot distribution of Academic Performance
plt.figure(figsize=(6, 4))
sns.countplot(data=Placement, x='Academic_Performance', order=sorted(Placement['Academic_Performance'].unique()), palette='magma')
plt.title("Academic Performance of Placed Students")
plt.xlabel("Academic Performance Score")
plt.ylabel("Number of Students")
plt.savefig("Academic_Performance2.png", dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
